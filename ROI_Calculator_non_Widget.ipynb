```python
!pip install numpy-financial
import numpy as np
import numpy_financial as npf
import pandas as pd

# ===== ÏûÖÎ†•Í∞í =====
Îß§ÏûÖÍ∞Ä = 320_000_000
LTVÎ™©Î°ù = [0.6, 0.7]
CapRateÎ™©Î°ù = [0.03, 0.04]
ÎåÄÏ∂úÏù¥ÏûêÏú® = 0.02
ÎåÄÏ∂úÍ∏∞Í∞Ñ = 25
Î≥¥Ïú†Í∏∞Í∞Ñ = 10
Î≤ïÏù∏ÏÑ∏Ïú® = 0.3
NOIÏÑ±Ïû•Î•† = 0.03
ExitCapRateÏ°∞Ï†ï = -0.005

Í≤∞Í≥º = []

for LTV in LTVÎ™©Î°ù:
    for CapRate in CapRateÎ™©Î°ù:
        ÎåÄÏ∂úÍ∏à = Îß§ÏûÖÍ∞Ä * LTV
        ÏûêÍ∏∞ÏûêÎ≥∏ = Îß§ÏûÖÍ∞Ä - ÎåÄÏ∂úÍ∏à
        NOI = Îß§ÏûÖÍ∞Ä * CapRate
        ExitCapRate = CapRate + ExitCapRateÏ°∞Ï†ï

        r = ÎåÄÏ∂úÏù¥ÏûêÏú®
        n = ÎåÄÏ∂úÍ∏∞Í∞Ñ
        Ïó∞Í∞ÑÏÉÅÌôòÏï° = ÎåÄÏ∂úÍ∏à * (r * (1 + r)**n) / ((1 + r)**n - 1)
        ÎåÄÏ∂úÏûîÏï° = ÎåÄÏ∂úÍ∏à

        ÌòÑÍ∏àÌùêÎ¶Ñ = [-ÏûêÍ∏∞ÏûêÎ≥∏]
        ÏÑ∏ÌõÑÏù¥ÏùµÎ™©Î°ù = []

        for year in range(1, Î≥¥Ïú†Í∏∞Í∞Ñ+1):
            if year > 1:
                NOI *= (1 + NOIÏÑ±Ïû•Î•†)
            Ïù¥Ïûê = ÎåÄÏ∂úÏûîÏï° * r
            ÏõêÍ∏à = Ïó∞Í∞ÑÏÉÅÌôòÏï° - Ïù¥Ïûê
            ÎåÄÏ∂úÏûîÏï° -= ÏõêÍ∏à
            Í≥ºÏÑ∏ÏÜåÎìù = NOI - Ïù¥Ïûê
            ÏÑ∏Í∏à = Í≥ºÏÑ∏ÏÜåÎìù * Î≤ïÏù∏ÏÑ∏Ïú®
            ÏÑ∏ÌõÑÏù¥Ïùµ = Í≥ºÏÑ∏ÏÜåÎìù - ÏÑ∏Í∏à
            Ïó∞Í∞ÑÌòÑÍ∏à = ÏÑ∏ÌõÑÏù¥Ïùµ - ÏõêÍ∏à
            ÏÑ∏ÌõÑÏù¥ÏùµÎ™©Î°ù.append(ÏÑ∏ÌõÑÏù¥Ïùµ)
            ÌòÑÍ∏àÌùêÎ¶Ñ.append(Ïó∞Í∞ÑÌòÑÍ∏à)

        Îß§Í∞ÅÍ∞Ä = NOI / ExitCapRate
        Îß§Í∞ÅÏ∞®Ïùµ = Îß§Í∞ÅÍ∞Ä - ÎåÄÏ∂úÏûîÏï°
        ÌòÑÍ∏àÌùêÎ¶Ñ[-1] += Îß§Í∞ÅÏ∞®Ïùµ

        IRR = npf.irr(ÌòÑÍ∏àÌùêÎ¶Ñ)

        Í≤∞Í≥º.append({
            "LTV": f"{int(LTV*100)}%",
            "Cap Rate": f"{int(CapRate*100)}%",
            "ÏûêÍ∏∞ÏûêÎ≥∏(Ïóî)": int(ÏûêÍ∏∞ÏûêÎ≥∏),
            "Exit ÏàòÏùµ(Ïóî)": int(Îß§Í∞ÅÏ∞®Ïùµ),
            "IRR": f"{IRR*100:.2f}%",
            "ÏÑ∏ÌõÑ Ïó∞ÌèâÍ∑† Î∞∞ÎãπÎ•†": f"{(np.mean(ÏÑ∏ÌõÑÏù¥ÏùµÎ™©Î°ù)/ÏûêÍ∏∞ÏûêÎ≥∏)*100:.2f}%"
        })

df = pd.DataFrame(Í≤∞Í≥º)
df
```
